```python
# 리스트와 반복 구조
#주어진 데이터가 어디에 저장되어 있는지 출력
data_list = [23, 45, 21, 14, 2, 67]
target = 14

index = 0
while index <= 5:
    if data_list[index] == target:
    break
    index += 1

print(index)
```

```python
for index in range(6):
    if data_list[index] == target:
        break

print(index)
```

```python
# 리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최대값)
data_list = [23, 45, 21, 14, 2, 67]

max = data_list[0]
max_index = 0
for index in range(1, 6):
    if data_list[index] > max:
        max = data_list[index]
        max_index = index

print(max, max_index)
```

```python
data_list = [123, 45, 10, 14, 2, 67]

max = data_list[0]
for data in data_list:
    if data > max:
        max = data

print(max)
```

```python
data_list = [23, 45, 21, 14, 2, 67]

min = data_list[0]
min_index = 0
for index in range(1, 6):
    if data_list[index] < min:
        min = data_list[index]
        min_index = index

print(min, min_index)
```

```python
data_list = [23, 45, 21, 14, 2, 67]
number = int(input('숫자를 입력하세요 :'))
GT = 0
for data in data_list:
    if data > number:
        GT += 1

print(GT)
```

```python
data_list = [23, 45, 21, 14, 2, 67]
sum = 0
for data in data_list:
    sum += data

print(sum)
```

```python
data_list = [23, 45, 21, 14, 2, 67]
swap_list = []


for data in data_list:
    swap_list.insert(0, data)


print(swap_list)
```
