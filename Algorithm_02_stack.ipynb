```python
# 데이터 정의
stack = []

while True:
# 메뉴를 출력 
    print(''' ==============================
 1.Push 2.Pop 3.Peek 0.Exit
 ==============================
    ''')
    # 입력을 받는다 
    menu = int(input('>> 연산 메뉴 선택 :'))
    #print(menu)
    
    # 입력에 따라 stack 연산을 수행하고 스택의 상태를 출력한다
    if menu == 0:
        break
    elif menu == 1: # push
        data = int(input('[Push] 추가할 데이터 입력 :'))
        stack.append(data)
    elif menu == 2: # pop
        data = stack.pop()
        print('[Pop] 가져온 데이터 :', data)
    elif menu == 3: #peak
        data = stack[len(stack)-1]
        print('top 데이터 확인 :', data)
        continue
    
print('[Stack]', stack)


print('정수형 스택 종료')
# 스택의 상태를 출력한다

# 위의 과정을 0. Exit을 선택할 때까지 반복
```

```python
# BOJ 10828
stack = []
N = int(input())
for _ in range(N):
    operation = input().split()
    if operation[0] == 'push':
        data = int(operation[1])
        stack.append(data)
    elif operation[0] == 'pop':
        if stack:
            data = stack.pop()
            print(data)
        else:
            print(-1)
    elif operation[0] == 'size':
        print(len(stack))
    elif operation[0] == 'empty':
        if not stack:
            print(1)
        else:
            print(0)
    elif operation[0] == 'top':
        if stack:
            print(stack[-1])
        else:
            print(-1)
```

```python
import sys
lst = []
q = int(input())
for i in range(q) :
    prs= sys.stdin.readline().split()
    cmd = prs[0]
    ln = len(lst)
    if cmd == 'push' :
        lst.append(prs[1])
    if cmd == 'pop' :
        if ln == 0 : 
            print(-1)
        else :
            print(lst.pop())
    if cmd == 'size' :
        print(ln)
    if cmd == 'empty' :
        if ln == 0 :
            print(1)
        else :
            print(0)
    if cmd == 'top' :
        if ln == 0 :
            print(-1)
        else :
            print(lst[ln - 1])
```
