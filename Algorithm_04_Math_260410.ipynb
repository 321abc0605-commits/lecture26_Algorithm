```python
# 정수 N을 받아서 base에 대한 진법의 수로 변환
def convert_base(N, base):
    # N을 base로 나눈 몫을 계속 N으로 나눈다. 몫이 0이 아닌 동안
    # 나눈 나머지를 진법에 해당하는 굿자를 변환하여 리스트에 저장한다.
    # N: 몫을 저장
    # new_list : 나머지를 저장하는 리스트
    new_number = ''
    digits = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ' #36진법까지..
    while N != 0:
        new_number = digits[N % base] + new_number
        N //= base
    return new_number


print(convert_base(25, 2))
```
